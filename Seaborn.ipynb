```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# **different Types of Graph**

Count Plot

```python
sns.countplot(x = "Sex", data = df)
## x- the name of the column
## data - the dataframe
```

```python
sns.countplot(x = "Sex", data = df, hue = "Pclass")
## hue - the name of the categorical column to split the bar
```

KDE Plot

```python
#KDE- Kernal Distribution Estimate
sns.kdeplot(x = "Age", data = df)
""" x = the name of the column
data- the dataset"""
```

```python
sns.kdeplot(x = "Age", data = df, color = "green")
```

Distribution Plot

```python
sns.displot(x ="Age", data = df)
```

```python
sns.displot(x ="Age", data = df, kde = True) # By default kde is false
```

```python
sns.displot(x ="Age", data = df, kde = True, bins = 5)
```

```python
sns.displot(x ="Age", data = df, kde = True, bins = 5, hue = df["Survived"])
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df2 = pd.read_csv("/content/12. tips.csv")
```

```python
df2.head()
```

```python
sns.set_style("dark")
sns.barplot(x = "sex", y = "total_bill", data = df2)
```

```python
sns.set_style("dark")
sns.barplot(x = "sex", y = "total_bill", data = df2, palette = "plasma")
```

```python
sns.boxplot(x = "day", y = "total_bill", data = df2)
```

```python
sns.violinplot(x ="day", y ="total_bill", data = df2)
```

```python
sns.set_style("dark")
sns.stripplot(x ="day", y ="total_bill", data = df2, jitter = True)
```

```python
sns.swarmplot(x ="day", y ="total_bill", data = df2, color = "Black")
```

```python
sns.swarmplot(x ="day", y ="total_bill", data = df2, color = "Black")
sns.violinplot(x ="day", y ="total_bill", data = df2)
```

```python
sns.catplot(x ="day", y ="total_bill", data = df2, kind = "bar")
```
